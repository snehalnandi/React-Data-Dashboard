```python
import pandas as pd
import numpy as np

# A messy inventory list where some quantities are missing (NaN)
messy_data = {
    'Item': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'Price': [50000, 800, 1500, 12000],
    'Stock_Quantity': [10, np.nan, 15, np.nan]
}

inventory_df = pd.DataFrame(messy_data)
print(inventory_df)
```

```python
# We tell Pandas: take the column, fill the holes with 0, and save it back!
inventory_df['Stock_Quantity'] = inventory_df['Stock_Quantity'].fillna(0)

# Print the fixed table
inventory_df
```

```python
reorder_items = inventory_df[ (inventory_df['Price'] > 1000) & (inventory_df['Stock_Quantity'] == 0) ]

print(reorder_items)
```

```python
inventory_df['Total_Value'] = inventory_df['Price'] * inventory_df['Stock_Quantity']
inventory_df
```

```python
inventory_df.to_csv('my_clean_inventory.csv', index=False)
```

```python
import pandas as pd
import numpy as np

cafe_data = {
    'Drink': ['Espresso', 'Cappuccino', 'Latte', 'Cold Brew', 'Mocha'],
    'Price': [100, 150, 180, 200, 220],
    'Cups_Sold': [50, np.nan, 30, np.nan, 40]
}

cafe_df = pd.DataFrame(cafe_data)

cafe_df
```

```python
cafe_df['Cups_Sold'] = cafe_df['Cups_Sold'].fillna(0)
cafe_df
```

```python
cafe_df['Total_Revenue'] = cafe_df['Price'] * cafe_df['Cups_Sold']
cafe_df
```

```python
Start_Drink = cafe_df [(cafe_df['Total_Revenue']>4000)&(cafe_df['Cups_Sold']>35)]
Start_Drink
```

```python
cafe_df.to_csv('star_drinks_report.csv',index=False)
```
